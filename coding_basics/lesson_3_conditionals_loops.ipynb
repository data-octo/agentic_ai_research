# Lesson 3: Conditional Statements and Loops
In this lesson, we will learn about conditional statements and loops in Python.

## Conditional Statements
Conditional statements allow us to make decisions in our code. For example, we can check if a withdrawal amount is greater than the account balance.

```python
# Example of a conditional statement
withdrawal_amount = 1200
if withdrawal_amount > account_balance:
    print("Insufficient funds for withdrawal.")
else:
    account_balance -= withdrawal_amount
    print("Withdrawal successful. New balance is:", account_balance)
```

## Loops
Loops allow us to perform repetitive tasks. For example, we can iterate through a list of transactions.

```python
# Example of a loop
transactions = [100, -50, 200, -30]
for transaction in transactions:
    account_balance += transaction
    print("Processed transaction:", transaction, "| New balance:", account_balance)
```
