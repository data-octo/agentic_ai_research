# Lesson 2: Variables and Data Types
In this lesson, we will learn about variables and data types in Python.

## What is a Variable?
A variable is a container for storing data. For example, you can use a variable to store the balance of a bank account.

```python
# Example of a variable
account_balance = 1000
print("The account balance is:", account_balance)
```

## Data Types
Python has different data types, such as numbers, strings, and lists.

```python
# Examples of data types
account_number = "123456789"  # String
interest_rate = 0.05  # Float
transactions = [100, -50, 200]  # List
print(account_number, interest_rate, transactions)
```
