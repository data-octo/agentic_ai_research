## Simple Banking System
Combine all the concepts to create a simple banking system with features like creating an account, depositing money, withdrawing money, and viewing account details.

```python
# Simple Banking System
def create_account(name, initial_balance):
    return {"name": name, "balance": initial_balance}
def deposit(account, amount):
    account["balance"] += amount
    print(f"Deposited ${amount}. New balance is: ${account['balance']}")
def withdraw(account, amount):
    if amount > account["balance"]:
        print("Insufficient funds for withdrawal.")
    else:
        account["balance"] -= amount
        print(f"Withdrew ${amount}. New balance is: ${account['balance']}")
def view_account(account):
    print("Account Details:")
    print("Name:", account["name"])
    print("Balance:", account["balance"])
# Example usage
my_account = create_account("Alice", 1000)
view_account(my_account)
deposit(my_account, 200)
withdraw(my_account, 500)
view_account(my_account)
```
