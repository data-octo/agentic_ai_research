## Variables and Data Types
Explain variables and data types in Python. Use examples like storing account balances and customer names.

```python
# Variables and Data Types
# Example: Storing account balances and customer names
customer_name = "Alice"
account_balance = 1000.50

print("Customer Name:", customer_name)
print("Account Balance:", account_balance)
```

## Basic Arithmetic Operations
Teach basic arithmetic operations like addition, subtraction, multiplication, and division. Use examples like calculating total balance after a deposit.

```python
# Basic Arithmetic Operations
# Example: Calculating total balance after a deposit
deposit_amount = 200
new_balance = account_balance + deposit_amount

print("After depositing $200, the new balance is:", new_balance)
```

## Conditional Statements
Introduce if-else statements to make decisions. Use examples like checking if a withdrawal amount is greater than the account balance.

```python
# Conditional Statements
# Example: Checking if a withdrawal amount is greater than the account balance
withdrawal_amount = 1200

if withdrawal_amount > account_balance:
    print("Insufficient funds for withdrawal.")
else:
    account_balance -= withdrawal_amount
    print("Withdrawal successful. New balance is:", account_balance)
```

## Loops and Iterations
Explain loops (for and while) to perform repetitive tasks. Use examples like iterating through a list of transactions.

```python
# Loops and Iterations
# Example: Iterating through a list of transactions
transactions = [100, -50, 200, -30]

for transaction in transactions:
    account_balance += transaction
    print("Processed transaction:", transaction, "| New balance:", account_balance)
```

## Functions and Reusability
Teach how to create and use functions. Use examples like creating a function to calculate interest on a savings account.

```python
# Functions and Reusability
# Example: Function to calculate interest on a savings account
def calculate_interest(balance, rate):
    return balance * rate / 100

interest = calculate_interest(account_balance, 5)
print("Interest on the current balance at 5% rate is:", interest)
```

## Lists and Dictionaries
Introduce lists and dictionaries to store multiple pieces of data. Use examples like storing customer details or transaction histories.

```python
# Lists and Dictionaries
# Example: Storing customer details and transaction histories
customers = [
    {"name": "Alice", "balance": 1000.50},
    {"name": "Bob", "balance": 500.75}
]

for customer in customers:
    print("Customer:", customer["name"], "| Balance:", customer["balance"])
```

## Simulating a Simple Banking System
Combine all the concepts to create a simple banking system. Include features like creating an account, depositing money, withdrawing money, and viewing account details.

```python
# Simulating a Simple Banking System
# Example: Simple banking system with basic features

def create_account(name, initial_balance):
    return {"name": name, "balance": initial_balance}

def deposit(account, amount):
    account["balance"] += amount
    print(f"Deposited ${amount}. New balance is: ${account['balance']}")

def withdraw(account, amount):
    if amount > account["balance"]:
        print("Insufficient funds for withdrawal.")
    else:
        account["balance"] -= amount
        print(f"Withdrew ${amount}. New balance is: ${account['balance']}")

def view_account(account):
    print("Account Details:")
    print("Name:", account["name"])
    print("Balance:", account["balance"])

# Example usage
my_account = create_account("Alice", 1000)
view_account(my_account)
deposit(my_account, 200)
withdraw(my_account, 500)
view_account(my_account)
```
